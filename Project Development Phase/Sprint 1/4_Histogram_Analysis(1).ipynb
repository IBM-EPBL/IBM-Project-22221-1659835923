```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
```

```python
A = imread('/content/7b0dc46f-4bfd-4329-85a3-06db12434d3a___RS_HL 7603.JPG')
B = imread('/content/7b236824-defb-40ae-8f83-76c00cf8b822___RS_HL 6222.JPG')
```

```python
plt.figure()
plt.subplot(121), plt.imshow(A)
plt.subplot(122), plt.imshow(B)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(np.abs(A[:, :, 0].astype(float) - B[:, :, 0].astype(float)), cmap='gray')
plt.show()
```

```python
d = imread('/content/7b0dc46f-4bfd-4329-85a3-06db12434d3a___RS_HL 7603.JPG')
mask = imread('/content/7b236824-defb-40ae-8f83-76c00cf8b822___RS_HL 6222.JPG')
print(np.amin(d), np.amax(d))
print(np.amin(mask), np.amax(mask))
```
